```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("CarPrice.csv")
```

```python
cars.head()
```

```python
cars.info()
```

```python
# symboling: -2 (least risky) to +3 most risky
# Most cars are 0,1,2
df=cars['symboling'].value_counts()
print(df
     )

sns.boxplot(y=cars['symboling'])
plt.title('symboling boxplot')

plt.show()



sns.distplot(cars['symboling'], hist=True)
plt.savefig('symboling1.png')
plt.show()



sns.barplot(x='symboling', y='price', data=cars, estimator=np.median)
plt.savefig('symboling2.png')
plt.show()
```

```python
df=(cars['aspiration'].astype('category').value_counts())
print(df)

df1=pd.DataFrame(df)

print(df1)

sns.barplot(x='aspiration', y='price', data=cars, estimator=np.median)
plt.title("median Price")
plt.savefig('aspiration.png')
plt.show()



```

```python

sns.barplot(x='aspiration', y='price', data=cars, estimator=np.mean)
plt.title("mean Price")
plt.show()
```

```python
 #drivewheel: frontwheel, rarewheel or four-wheel drive 
cars['drivewheel'].astype('category').value_counts()
```

```python

sns.barplot(x='drivewheel', y='price', data=cars, estimator=np.mean)
plt.title("mean Price")
plt.savefig('drivewheel.png')
plt.show()
```

```python
# curbweight: weight of car without occupants or baggage
sns.distplot(cars['curbweight'])
plt.savefig('curbweight.png')
plt.show()
```

```python

sns.jointplot('curbweight', 'price', cars)
plt.savefig('curbweight1.png')
plt.show()
```

```python
# stroke: volume of the engine (the distance traveled by the 
# piston in each cycle)
sns.distplot(cars['stroke'])
plt.savefig('stroke.png')
plt.show()
```

```python
# compression ration: ratio of volume of compression chamber 
# at largest capacity to least capacity
sns.distplot(cars['compressionratio'])
plt.savefig('compressionratio.png')
plt.show()
```

```python
# target variable: price of car
sns.distplot(cars['price'])
plt.show()

sns.boxplot(y=cars['price'])
plt.title('price boxplot')
plt.savefig('price.png')
plt.show()
```

```python
sns.pairplot(cars)
plt.show()
```

```python
cor=cars.corr()
cor
```

```python
plt.figure(figsize = (16,8))
sns.heatmap(cor,cmap="YlGnBu",annot=True)
plt.savefig('price_heatmap.png')
plt.show()
```
