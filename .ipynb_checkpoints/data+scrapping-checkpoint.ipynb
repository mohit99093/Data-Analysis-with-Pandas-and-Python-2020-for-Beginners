### Scrapping data from websites
#### data access from websites can give the deep insights like customer reviews , feedback on the products etc.

#### we will learn how to extract meaningful data from websites

```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.monster.com/jobs/search/?q=Software-Developer&where=India'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')

soup
```

```python
results = soup.find(id='SearchResults')
results
```

```python
jobs = results.find_all('section', class_='card-content')
jobs
```

```python
for i in jobs:
    print(jobs, end='\n'*2)
```

```python
for i in jobs:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = i.find('h2', class_='title')
    company_elem = i.find('div', class_='company')
    location_elem = i.find('div', class_='location')
    print(title_elem)
    print(company_elem)
    print(location_elem)
    print()
```

```python
import PyPDF2

# reading the pdf file
pdf_object = open('anaconda intallation.pdf', 'rb')
pdf_reader = PyPDF2.PdfFileReader(pdf_object)

# Number of pages in the PDF file
print(pdf_reader.numPages)

# get a certain page's text
page_object = pdf_reader.getPage(4)

# Extract text from the page_object
print(page_object.extractText())
```
