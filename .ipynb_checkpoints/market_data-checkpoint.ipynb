### Merging multiple files or dataframes

```python
import numpy as np
import pandas as pd

market_fact = pd.read_csv("market_fact.csv")
cust_dimen = pd.read_csv("cust_dimen.csv")
prod_dimen = pd.read_csv("prod_dimen.csv")
shipping_dimen = pd.read_csv("shipping_dimen.csv")
orders_dimen = pd.read_csv("orders_dimen.csv")
```

```python
market_fact.head()
```

```python
prod_dimen.head()
```

```python
shipping_dimen.head()
```

```python
orders_dimen.head()
```

```python
market_cust = pd.merge(market_fact, cust_dimen, how='inner', on='Cust_id')
market_cust.head()
```

```python
#pay attention to this code , which is like a filter condition
market_cust.loc[market_cust['Ord_id'] == 'Ord_5446', :]
```

```python
# merging product DF now.
market_cust_prod = pd.merge(market_cust, prod_dimen, how='inner', on='Prod_id')

market_cust_prod.head()
```

```python
market_cust_prod.loc[(market_cust_prod['Product_Category']=='OFFICE SUPPLIES') & (market_cust_prod['Customer_Segment']=='CORPORATE'),:]
```

```python
# Merging the orders table
market_cust_prod_order = pd.merge(market_cust_prod, orders_dimen, how='inner', on='Ord_id')
market_cust_prod_order.shape
market_cust_prod_order.head()
```

```python
# dataframes having the same columns
df1 = pd.DataFrame({'Name': ['John', 'Tom', 'Mohan', 'Kel'],
                    'Age': ['34', '31', '22', '33'],
                    'Gender': ['M', 'M', 'M', 'F']}
                  )

df2 = pd.DataFrame({'Name': ['Anson', 'Ruby', 'Mai'],
                    'Age': ['31', '22', '23'],
                    'Gender': ['M', 'F', 'F']}
                  )
print(df1)
print('/n')
print(df2)
```

```python
# To concatenate them, one on top of the other, you can use pd.concat
# The first argument is a sequence (list) of dataframes
# axis = 0 indicates that we want to concat along the row axis
pd.concat([df1, df2], axis = 0)
```

```python
# To join the two dataframes, use axis = 1 to indicate joining along the columns axis
# The join is possible because the corresponding rows have the same indices
pd.concat([df1, df2], axis = 1)
```

### arithmetic operations that can be performed on dataframes are as follow:

add(): +

sub(): -

mul(): *

div(): /

floordiv(): //

mod(): %

pow(): **

```python
FIFA_2018 = pd.DataFrame({'FIFA_TEAM': ['BEL', 'FRANCE', 'CROTIA', 'JAPAN', 'SWEDEN', 'BRAZIL'],
                         'Matches Played': [7, 7, 7, 4, 5, 7],
                         'Goals Scored': [11, 10, 9, 7, 6, 6]}
                       )

# Set the 'IPL Team' column as the index to perform arithmetic operations on the other rows using the team as reference
FIFA_2018.set_index('FIFA_TEAM', inplace = True)
FIFA_2018
```

```python
FIFA_2014 = pd.DataFrame({'FIFA_TEAM': ['BEL', 'FRANCE', 'CROTIA', 'JAPAN', 'SWEDEN', 'BRAZIL','KOREA'],
                         'Matches Played': [7, 6, 7, 4, 5, 7,10],
                         'Goals Scored': [8, 7, 9, 10, 6, 5,3]}
                       )

# Set the 'IPL Team' column as the index to perform arithmetic operations on the other rows using the team as reference
FIFA_2014.set_index('FIFA_TEAM', inplace = True)
FIFA_2014
```

```python
Total = FIFA_2014 + FIFA_2018
Total
```

```python
# The fill_value argument inside the df.add() replaces the NaN value with 0
Total = FIFA_2018.add(FIFA_2014, fill_value = 0)
Total
```

```python
# Creating a new column - 'Win Percentage'

Total['Goals Percentage'] = Total['Goals Scored']/Total['Matches Played']
Total
```

```python

Total.sort_values(by = (['Goals Scored', 'Goals Percentage']), ascending = False)
```

### Grouping  , Aggregators 

For any raw data , to be processed further , the data engineer must perform the data analysis . This is the most important task in data science programming. 

```python
# Merging the orders table
market_cust_prod_order = pd.merge(market_cust_prod, orders_dimen, how='inner', on='Ord_id')
market_cust_prod_order.shape
market_cust_prod_order.head()
```

```python
mcpo_df=market_cust_prod_order
mcpo_df.head()
```

```python
# Which customer segments are the least profitable? 

# Step 1. Grouping: First, we will group the dataframe by customer segments
mcpo_by_segment = mcpo_df.groupby('Customer_Segment')
mcpo_by_segment.size()
```

```python
# applying the function of sum
mcpo_by_segment['Profit'].sum()
```

```python
# Alternatively
mcpo_by_segment.Profit.sum()
```

```python
# For better readability, you may want to sort the summarised series:
mcpo_by_segment.Profit.sum().sort_values(ascending = False)
```

### Summarising Dataframes

```python
## we have imported the following file in the previous sessions

import numpy as np
import pandas as pd

market_fact = pd.read_csv("market_fact.csv")
cust_dimen = pd.read_csv("cust_dimen.csv")
prod_dimen = pd.read_csv("prod_dimen.csv")
shipping_dimen = pd.read_csv("shipping_dimen.csv")
orders_dimen = pd.read_csv("orders_dimen.csv")
```

```python
market_fact.head()
```

```python
cust_dimen.head()
```

```python
# Merging the dataframes one by one
mar_cust = pd.merge(market_fact, cust_dimen, how='inner', on='Cust_id')
mar_cust_pr = pd.merge(mar_cust, prod_dimen, how='inner', on='Prod_id')
mar_cust_pr_ship = pd.merge(mar_cust_pr, shipping_dimen, how='inner', on='Ship_id')
mar_cust_pr_ship_ord = pd.merge(mar_cust_pr_ship, orders_dimen, how='inner', on='Ord_id')

mar_cust_pr_ship_ord.head()
```

```python
# which product category has the highest sales 

group_prod_cat = mar_cust_pr_ship_ord.groupby('Product_Category')
group_prod_cat.size()
```

```python
group_prod_cat['Sales'].sum()
```

```python
pd.DataFrame(group_prod_cat.Sales.sum())
```

```python
pd.DataFrame(group_prod_cat.Sales.sum().sort_values(ascending = False))
```

```python
# which product category and sub-category has the highest sales 
group_prod_sub_cat = mar_cust_pr_ship_ord.groupby(['Product_Category','Product_Sub_Category'])
group_prod_sub_cat.size()
```

```python
group_prod_sub_cat.Sales.sum()
pd.DataFrame(group_prod_sub_cat.Sales.sum())
```

```python
#make the use of describe fucntion to have overall statistics of data
group_prod_sub_cat.Sales.describe()
```

```python
sum(mar_cust_pr_ship_ord.Sales)
```

```python
#sales % 
group_prod_sub_cat.Sales.sum()/sum(mar_cust_pr_ship_ord.Sales)*100
```

### Using Lambda Function to modify the current data in columns or add new columns
### Using Pivot function as an alternative to Group by 

```python
# using apply () function , return the values 'Y' if the sales > 500 else return 'N'

def sales_500(x):
    if x>500:
     return True
    elif x<=500:
     return False
       

```

```python
# Create a new column and using the apply function
mar_cust_pr_ship_ord['sales_500'] = mar_cust_pr_ship_ord['Sales'].apply(sales_500)
mar_cust_pr_ship_ord.head()
```

```python
#use the lambda function ( here no need to create another df function)
mar_cust_pr_ship_ord['sales_500'] = mar_cust_pr_ship_ord['Sales'].apply(lambda x: True if x > 500 else False )
mar_cust_pr_ship_ord.head()
```

```python
#use the lambda function modify the existing data 
#using round function to remove decimals
mar_cust_pr_ship_ord['Sales'] = mar_cust_pr_ship_ord['Sales'].apply(lambda x: round(x) )
mar_cust_pr_ship_ord.head()
```

### PIVOT TABLES

pivot functions prorvides excel like functionality 

the syntax of pivot table pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', ...).

```python
#mar_cust_pr_ship_ord.pivot_table(values = 'Sales', index = 'Product_Category', aggfunc = 'mean')
mar_cust_pr_ship_ord.pivot_table(values = 'Sales', index = 'Product_Category', aggfunc = 'max')
```

```python
#columns / rows pivoting
mar_cust_pr_ship_ord.pivot_table(values = 'Sales', 
                      index = 'Product_Category', 
                      columns = 'Product_Sub_Category', 
                      aggfunc = 'sum')
```
