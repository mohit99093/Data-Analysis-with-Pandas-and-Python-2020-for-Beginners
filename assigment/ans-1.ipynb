```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import datetime

import seaborn as sns

data = pd.read_csv("Covid19_data_mar.csv")

data1=data.drop("exposure_start",axis =1)

data1=data1.drop("exposure_end",axis =1)

data1=data1.drop("hosp_visit_date",axis =1)

data1=data1.drop("death",axis =1)

data1=data1.drop("recovered",axis =1)

data1=data1[~np.isnan(data1['case_in_country'])]

data1=data1[~np.isnan(data1['If_onset_approximated'])]

data1=data1[~np.isnan(data1['age'])]

print(round(data1.isnull().sum()/len(data1.index)*100),2)
```
