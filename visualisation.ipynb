```python
# reading new sets of files crptocurrency files where we will study the relationshop among different prices of this currency
btc = pd.read_csv("bitcoin_price.csv")
ether = pd.read_csv("ethereum_price.csv")
ltc = pd.read_csv("litecoin_price.csv")
monero = pd.read_csv("monero_price.csv")
neo = pd.read_csv("neo_price.csv")
quantum = pd.read_csv("qtum_price.csv")
ripple = pd.read_csv("ripple_price.csv")

#suffixing  column names so that joins are easy
btc.columns = btc.columns.map(lambda x: str(x) + '_btc')
ether.columns = ether.columns.map(lambda x: str(x) + '_et')
ltc.columns = ltc.columns.map(lambda x: str(x) + '_ltc')
monero.columns = monero.columns.map(lambda x: str(x) + '_mon')
neo.columns = neo.columns.map(lambda x: str(x) + '_neo')
quantum.columns = quantum.columns.map(lambda x: str(x) + '_qt')
ripple.columns = ripple.columns.map(lambda x: str(x) + '_rip')

btc.head()
```

```python
# merging all the files by date
m1 = pd.merge(btc, ether, how="inner", left_on="Date_btc", right_on="Date_et")
m2 = pd.merge(m1, ltc, how="inner", left_on="Date_btc", right_on="Date_ltc")
m3 = pd.merge(m2, monero, how="inner", left_on="Date_btc", right_on="Date_mon")
m4 = pd.merge(m3, neo, how="inner", left_on="Date_btc", right_on="Date_neo")
m5 = pd.merge(m4, quantum, how="inner", left_on="Date_btc", right_on="Date_qt")
crypto = pd.merge(m5, ripple, how="inner", left_on="Date_btc", right_on="Date_rip")

crypto.head()
```

```python
# Subsetting only the closing prices column for plotting
curr = crypto[["Close_btc", "Close_et", 'Close_ltc', "Close_mon", "Close_neo", "Close_qt"]]
curr.head()
```

```python
# pairplot
sns.pairplot(curr)
plt.show()
```

```python
# You can also observe the correlation between the currencies 
# using df.corr()
cor = curr.corr()
round(cor, 3)
```

```python
# figure size
#plt.figure(figsize=(10,8))

plt.figure(num=None, figsize=(12, 8))

# heatmap

sns.heatmap(cor, cmap="YlGnBu", annot=True)

plt.savefig('heatmap1.png')
plt.show()
```

## Plotting Time Series Data

```python
# merging market fact data with order dimension as it has data value  ( order_date)

mark_ord = pd.merge(market_fact, orders_dimen, how='inner', on='Ord_id')
mark_ord.head()
```

```python
mark_ord.info()

# order date here is string
```

```python
mark_ord['Order_Date'] = pd.to_datetime(mark_ord['Order_Date'])
mark_ord.info()

```

```python
## grouping the sale values by date 

date_sales = mark_ord.groupby('Order_Date')['Sales'].sum()
print(date_sales.head())

print(type(date_sales))
date_sales
```

```python
# to plot time series plot we will use tsplot 

plt.figure(figsize=(16, 8))

sns.lineplot(data=date_sales)
plt.show()

# note that x-axis does not show dates , you need to further analyse on how to show the dates.
```

```python
# extracting month and year from date

# extract month
mark_ord['month'] = mark_ord['Order_Date'].dt.month

# extract year
mark_ord['year'] = mark_ord['Order_Date'].dt.year

mark_ord.head()
```

```python
data_year_mon = mark_ord.groupby(["year", "month"]).Sales.mean()
data_year_mon.head()
```

```python
plt.figure(figsize=(8, 6))
# time series plot
sns.lineplot(data=data_year_mon)
plt.xlabel("year month")
plt.ylabel("Mean Sales")
plt.savefig('timeseries.png')
plt.show()
```

```python
# Pivoting the data using 'month' 
data_year_mon1 = pd.pivot_table(mark_ord, values='Sales', index='year', columns='month', aggfunc='mean')
data_year_mon1.head()
```

```python
# figure size
plt.figure(figsize=(12, 8))

# heatmap with a color map of choice
sns.heatmap(data_year_mon1, cmap="YlGnBu")
plt.show()
```
